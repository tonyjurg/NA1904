# 1 - Introduction <a class="anchor" id="first-bullet"></a>

Greek has many particles, such as [μέν](https://stephanus.tlg.uci.edu/lsj/#eid=68439) and [δέ](https://stephanus.tlg.uci.edu/lsj/#eid=24443), that are used to indicate contrast or emphasis. 

## 1.1 - Why is this relevant? <a class="anchor" id="#bullet1_1"></a>

The particle δέ can be used in either an adversative (contrasting) or copulative (confirming) manner. These particles can be difficult to translate into English, and their meaning can be ambiguous.
